```python
#귀무가설

import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt

xx1 = np.linspace(-4, 4, 100)
black = {"facecolor" : "black"}
plt.figure(figsize = (8,4))
plt.subplot(121)
plt.plot(xx1, sp.stats.norm.pdf(xx1))
plt.plot(0.5,0,"ro")
plt.annotate('real test statistics', xy=(0.5,0.01), xytext = (0.85,0.1), arrowprops = black)
plt.subplot(122)
plt.plot(xx1, sp.stats.norm.pdf(xx1))
plt.plot(2.2,0,"ro")
plt.annotate('real test statistics $t_0$', xy=(2.2,0.01), xytext = (0.85,0.1), arrowprops = black)
plt.suptitle('asdasdasd',  y =1.05)
plt.tight_layout()
plt.show()
```

```python
#플롯
xx1 = np.linspace(-4, 4, 100) 
black = {"facecolor": "black"} 
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.title("Case of a high p-value")
plt.plot(xx1, sp.stats.norm.pdf(xx1))
plt.plot(0.5, 0, "ro")
plt.annotate('real test statistics $t_0$', xy = (0.5, 0.01), xytext = (0.85, 0.1), arrowprops = black)
xx2 = np.linspace(-4, -0.5, 100)
xx3 = np.linspace(0.5, 4, 100)
plt.fill_between (xx2, sp.stats.norm.pdf(xx2), facecolor='blue', alpha=0.35)
plt.fill_between (xx3, sp.stats.norm.pdf(xx3), facecolor='blue', alpha=0.35)
plt.annotate('p-value', xy=(-1.5, 0.05), xytext = (-3.5,0.1), arrowprops = black)
plt.subplot(122)
plt.title("Case of a low p-value")
plt.plot(xx1, sp.stats.norm.pdf(xx1))
plt.plot(2.2, 0, "ro")
plt.annotate('real test statistic $t_0$', xy=(2.2, 0.01), xytext=(0.85, 0.1), arrowprops=black) 
xx2 = np.linspace(-4, -2.2, 100)
xx3 = np.linspace (2.2, 4, 100)
plt.fill_between(xx2, sp.stats.norm.pdf(xx2), facecolor='blue', alpha=0.35)
plt.fill_between(xx3, sp.stats.norm.pdf(xx3), facecolor='blue', alpha=0.35)
plt.annotate('p-value', xy=(-2.5,0.01), xytext = (-3.5,0.1), arrowprops = black)
plt.suptitle("Dist. of the test statistics and the value of the real test statistics $t_0$ 의 값", y=1.05)
plt.tight_layout()
plt.show()
```

```python
N = 15
mu = 0.5
rv = sp.stats.binom(N, mu) 
xx = np.arange(N + 1) 
plt.subplot(211)
plt.stem(xx, rv.pmf(xx))
plt.ylabel("pmf")
plt.title("pmf of test statistics dist.(binomial diat. with N=15)")
black = {"facecolor": "black"}
plt.annotate('test statistics t=12',xy = (12, 0.02), xytext = (12, 0.1), arrowprops = black)
plt.subplot(212)
plt.stem(xx, rv.cdf(xx))
plt.ylabel("cdf")
plt.title("cdf of test statistics dist.(binomial diat. with N=15)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

rv= sp.stats.norm()
xx= np.linspace(-4,4,100)

plt.subplot(211)
plt.plot(xx, rv.pdf(xx))
plt.ylabel("pdf")
black = {"facecolor" : "black"}
plt.annotate('test statistics t = -1.4025', xy=(-1.4,0.15), xytext = (4,0.25), arrowprops = black)
xx2 = np.linspace(-4, -1.4025, 100)
plt.fill_between(xx2, rv.pdf(xx2), facecolor = 'blue', alpha = 0.35)

plt.subplot(212)
plt.plot(xx, rv.cdf(xx))
black = {"facecolor" : "black"}
plt.fill_between(xx2, rv.cdf(xx2), facecolor = 'black', alpha = 0.35)
plt.ylabel("cdf")
plt.tight_layout()
plt.show()
```
