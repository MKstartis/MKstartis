```python
(1) Bin(3, θ) 분포의 확률 질량 함수는 다음과 같습니다.

P(X = x) = comb(3, x) * θ^x * (1-θ)^(3-x)

여기서 comb(a, b)는 a개에서 b개를 선택하는 조합의 수를 나타냅니다.

따라서 데이터 (x1, x2, x3, x4) = (, 3, 2, 2)에 대한 가능도 함수는 다음과 같습니다.

L(θ|x1,x2,x3,x4) = P(X = 1) * P(X = 3) * P(X = 2) * P(X = 2)
                  = (3C1 * θ^1 * (1-θ2) * (3C3 * θ^3 * (1-θ)^0) 
                    * (3C2 * θ^2 * (1-θ)^1) * (3C2 * θ^2 * (1-θ)^1)
                  = 3θ(1-θ)^2 * θ^ * 1 * 3(θ^2)(1-θ) * 3(θ^2)(1-θ)
                  = 27θ^8 * (1 - θ)^6

() 최대 가능도 추정(MLE)은 가능도를 최대화하는 모수 값을 찾는 것입니다. 따라서 가능도 함수를 θ로 미분하고 미분이 0이 되는 θ 값을 구하면 됩니다.

L(θ|x1,x2,x3,x4) = 27θ^8 * (1 - θ)^6
log(θ|x1,x2,x3,x4) = 8log(θ) + 6log(1 - θ)
d/dθ logL(θ|x1,x2,x3,x4) = 8/θ - 6/(1 - θ)
                        = (8 - 6θ) / θ(1-θ)

d/dθ logL(θ|x1,x2,x3,x4) =0 의 조건을 풀면,

8 - 6θ = 0
θ = 4/6 = 2/3

따라서 x1, x2, x3, x4의 데이터로 인해 추정된 모 θ는 2/3입니다.
```
