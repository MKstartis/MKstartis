```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
linestytles = ['-','--',':']
```

```python
#베타분포✱ 감마분포✱ 디리클레분포는 모숫값을 조정하여 분포의 모양을 우리가 원하는대로 쉽게 바꿀 수 있다✳

fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111)

xs = np.linspace(0,3,100)
for lam, ls in zip([1,2,3], linestytles):
    rv = stats.expon(scale = 1/lam)
    ax.plot(xs, rv.pdf(xs),label=f'lambda:{lam}', ls= ls, color= 'gray')
ax.legend()

plt.show
```

```python
import numpy as np
import scipy as sp
import scipy.stats
#베타분포
xx= np.linspace(0,1,1000)
plt.subplot(221)
plt.fill_between(xx, sp.stats.beta(0.5,2).pdf(xx))
plt.ylim(0,6)
plt.title("(A) a= 1, b = 1")
plt.subplot(222)
plt.fill_between(xx, sp.stats.beta(5,2).pdf(xx))
plt.ylim(0,6)
plt.title("(B) a= 4, b = 2, mode = {0}".format((4-1)/(4+2-2)))
plt.subplot(223)
plt.fill_between(xx, sp.stats.beta(8,4).pdf(xx))
plt.ylim(0,6)
plt.title("(C) a= 8, b = 4, mode = {0}".format((8-1)/(8+4-2)))
plt.subplot(224)
plt.fill_between(xx, sp.stats.beta(30,12).pdf(xx))
plt.ylim(0,6)
plt.title("(D) a= 30, b = 122, mode = {0}".format((30-1)/(30+12-2)))
```

```python
#디리클레분포 응용
import seaborn as sns
def plot_triangle(X, kind):
    n1 = np.array([1,0,0])
    n2 = np.array([0,1,0])
    n3 = np.array([0,0,1])
    n12 = (n1+n2)/2
    m1 = np.array([1,-1,0])
    m2 = n3-n12
    m1 = m1/ np.linalg.norm(m1)
    m2 = m2/ np.linalg.norm(m2)
    
    X1= (X - n12).dot(m1)
    X2= (X - n12).dot(m2)
    
    sns.jointplot(X1,X2, kind=kind, xlim =(-0.8,0.8), ylim = (-0.1,1.25))
    plt.show()
    
```

```python
np.random.seed(0)
X1 = np.random.rand(1000,3)
X1 = X1/X1.sum(axis=1)[:,np.newaxis]
plot_triangle(X1, kind="scatter")
```
