```python
import scipy as sp
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def plot_binom(mu, N,M):
    rv=  sp.stats.binom(N,mu)
    x= rv.rvs(M, random_state = 0)
    y= np.bincount(x, minlength = N+1)/float(len(x))
    xx = np.arange(N+1)
    df= pd.DataFrame({'Theory': rv.pmf(xx),'Simulation': y}).stack()
    df= df.reset_index()
    df.columns = ['sample value', 'type','ratio']
    df.pivot('sample value', 'type','ratio')
    sns.barplot(x= 'sample value',y= 'ratio' , hue = 'type', data = df)
    plt.title('$\mu$={},$N$ = {}, smple_size = {}'.format(mu,N,M))
    plt.ylim(0,0.7)
```

```python
plt.subplot(221) #여러개의 플롯을 작성 3자리수의 연속한 정수를 입력받는 방식
plot_binom(0.5,5,10)
plt.subplot(222)
plot_binom(0.5,5,1000)
plt.subplot(223)
plot_binom(0.9,20,10)
plt.subplot(224)
plot_binom(0.9,20,1000)

plt.tight_layout()
plt.show()
```
